### 2. Retorno vs print
#### 5 - Crie uma função que retorne o quadrado de um número.
#### 6 - Refaça o exercício anterior usando print e explique a diferença.
#### 7 - Crie uma função que calcule a média de três números e retorne o valor.

```python
def quadrado(num: int|float) -> int|float:
    """Função que retorna o quadrado do número inserido."""
    return num ** 2

square1 = quadrado(2.2)
print(square1), type(square1)
```

```python
def quadrado_nula(num: int|float) -> int|float:
    """Função que apenas exibe o resultado do quadrado."""
    print(num ** 2)

square2 = quadrado_nula(2)
print(quadrado_nula), type(quadrado_nula)
```

```python
def calculo(x: int|float, c: int|float) -> int|float:
    """Função que calcula a soma de """
    sqr = quadrado(x)
    if x + c < sqr:
        print(f"A soma de {x} + {c} é menor do que o quadrado de {x}: {sqr}")
    else:
        print(f"A soma de {x} + {c} é maior do que o quadrado de {x}: {sqr}")

calculo(4,20)
```

```python
def calculo(x: int|float, c: int|float) -> int|float:
    """Função que verifica se a soma dos números x + c são menores do que o quadrado de x."""
    sqr = quadrado_nula(x)
    if x + c < sqr:
        print(f"A soma de {x} + {c} é menor do que o quadrado de {x}: {sqr}")
    else:
        print(f"A soma de {x} + {c} é maior do que o quadrado de {x}: {sqr}")

calculo(4,20)
```

##### Resposta minha: As funções nulas (com print) não retornam nenhum valor (valor de retorno = None). O valor exibido por elas não é armazenado na memória da função, fato que impossibilita o reuso do resultado da função em outros trechos do código. Em contrapartida, é possível reutilizar a função nula em outros trechos do código, o que não é possível é reutilizar o resultado de uma função nula em outros trechos, pois seu valor é None.

##### As funções com resultado (com return) armazenam o valor na memória da função, fato que possibilita reutilizar o resultado da função em outros trechos do código.

```python
def media(a: int|float,b: int|float,c: int|float) -> int|float:
    rol = a,b,c
    return sum(rol) / len(rol)

media(1,2,3)
```

## 3 - Parâmetros e argumentos

#### 8. Crie uma função que receba base e altura e calcule a área de um retângulo.

#### 9. Crie uma função que receba um número e um expoente (default = 2).

#### 10. Crie uma função que receba nome e idade e retorne uma frase formatada.

```python
def area_retangulo(b: int|float, h: int|float) -> int|float:
    """Função que retorna a área de um triângulo.
    b = base
    h = altura
    """
    area = b * h
    return area

area_retangulo(2.7, 10.6)
```

```python
def calculo2(num: int|float, exp:int|float=2) -> float|int:
    """Função que recebe um número e o eleva a um expoente.
    O expoente por padrão é o número 2.
    - num: número a ser elevado
    - exp: potência que elevará o número
    """
    return num ** exp

calculo2(2,3)
```

```python
def apresentacao(nome: str, idade:int|str) -> str:
    """Função de apresentação, recebe o nome, a idade e devolve uma frase de apresentação."""
    frase = f"Olá! Eu me chamo {nome} e tenho {idade} anos!"
    return frase

apresentacao('Daniel Ortega', 27)
```


### Nível 2 - Condições e Validações
#### 4 - Funções com if/else
##### 11. Crie uma função que diga se um número é par ou ímpar
##### 12. Crie uma função que diga se um número é positivo, negativo ou zero
#### 13. Crie uma função que receba uma nota e retorne: Aprovado se >= 7; Recuperação se entre 5 e 6.9; Reprovado se < 5 

```python
def par_impar(x:int|float) -> str:
    """Função que verifica um número e diz se ele é par ou ímpar."""
    if x % 2 == 0:
        print(f'O número {x} é par')
    else:
        print(f'O número {x} é ímpar')

par_impar(2.6)
```

```python
def meta_num(x:int|float):
    """Função que verifica se um número é positivo, negativo ou """
    if x > 0:
        print(f'O número {x} é positivo!')
    elif x == 0:
        print(f'O número {x} é igual a zero!')
    elif x < 0:
        print(f'O número {x} é negativo!')

meta_num(-1.9)
```

```python
def status_aluno(nota: int|float) -> str:
    """Função que verifica a nota de um aluno e retorna 
    se ele foi aprovado, reprovado ou está em recuperação."""
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota <= 6.9:
        return "Recuperação"
    elif nota < 5:
        return "Reprovado"
    
status_aluno(6.9)
```

```python
def so_positivo(numero:int|float) -> int|float|str:
    if numero > 0:
        return numero
    else:
        return "A função aceita apenas números positivos!"
    
so_positivo(-1.4)
```

```python
def tratamento_divisao(numerador: int|float, denominador: int|float):
    if numerador == 0:
        return 0
    elif denominador == 0:
        return 0
    elif numerador and denominador == 0:
        return 0
    else:
        return numerador / denominador
    
tratamento_divisao(0,0)
```

```python
def capitalize_all(t: str) -> str:
    res = [s.capitalize() for s in t]
    res.append(s)
    return res

capitalize_all("fluminense")
```
